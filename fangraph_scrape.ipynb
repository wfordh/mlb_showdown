```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
url = 'https://www.fangraphs.com/leaders.aspx?pos=all&stats=pit&lg=all&qual=y&type=8&season=2018&month=0&season1=2018&ind=0'
r = requests.get(url)
```

```python
c = r.content
```

```python
soup = BeautifulSoup(c, 'html.parser')
```

```python
data_rows = soup.find_all('tr')[1:]
```

```python
table = soup.find('table', {'class':'rgMasterTable'})
```

```python
[td.get_text() for td in table.find_all('tr')[3:6]]
```

```python
rows = table.find('tbody').find_all('tr')

data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols[1:]])
```
